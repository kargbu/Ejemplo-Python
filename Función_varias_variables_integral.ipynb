```python
# symbols para usar s√≠mbolos matem√°ticso, Function es definir
# funciones y calcular derivadas, solve es para escribir
# la soluci√≥n, Matriz para operar matrices, hessian es para
# representar la segunda derivada.

# plot3d es para vizualizar funciones en 3D


from sympy import symbols, Function, Eq, diff, solve, Matrix, hessian, sympify
from sympy.plotting import plot3d

x, y = symbols ('x y')
f = Function ('f')(x, y)
f

Expresion = x**2 + y**2
Eq(f, Expresion)

plot3d(Expresion, (x, -5, 5), (y, -5, 5))
```
