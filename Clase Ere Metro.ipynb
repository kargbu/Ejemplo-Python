```python
import pandas as pd
import numpy as np
import networkx as nx # es la librer√≠a para todo lo referente a gr√°ficas
```

```python
df = pd.read_excel('/content/metro.xlsx')
df.head()
```

```python
df.shape
```

```python
METRO = nx.from_pandas_edgelist(df,source='Origen',target='Destino',edge_attr='Longitud de interestaci√≥n')
METRO
```

```python
METRO.edges()
```

```python
djk_path= nx.dijkstra_path(METRO, source='Ciudad Azteca', target='Universidad', weight='Longitud de interestaci√≥n')
djk_path
```
