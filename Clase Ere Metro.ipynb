# Algoritmo de Dijkstra

Este algoritmo fue creado por uno de los padres de la computación, Edger W. Dijkstra, en 1956. Sirve para cualquier grafo con pesos (dirigido o no) siempre y cuando sus pesos no sean negativos.



images.jpg


```python
import pandas as pd
import numpy as np
import networkx as nx # es la librería para todo lo referente a gráficas
```

https://metro.cdmx.gob.mx/longitud-de-estacion


```python
df = pd.read_excel('/content/metro.xlsx')
df.head()
```

```python
df.shape
```

```python
METRO = nx.from_pandas_edgelist(df,source='Origen',target='Destino',edge_attr='Longitud de interestación')
METRO
```

```python
METRO.edges()
```

```python
djk_path= nx.dijkstra_path(METRO, source='Ciudad Azteca', target='Universidad', weight='Longitud de interestación')
djk_path
```

```python
ruta =METRO.subgraph(['Ciudad Azteca',
 'Plaza Aragón',
 'Olímpica',
 'Ecatepec',
 'Múzquiz',
 'Río de los Remedios',
 'Impulsora',
 'Nezahualcóyotl',
 'Villa de Aragón',
 'Bosques de Aragón',
 'Deportivo Oceanía',
 'Oceanía',
 'Romero Rubio',
 'Ricardo Flores Magón',
 'San Lázaro',
 'Candelaria',
 'Merced',
 'Pino Suárez',
 'Isabel la Católica',
 'Salto del Agua',
 'Balderas',
 'Niños Héroes',
 'Hospital General',
 'Centro Médico',
 'Etiopía/Plaza de la Transparencia',
 'Eugenia',
 'División del Norte',
 'Zapata',
 'Coyoacán',
 'Viveros/Derechos Humanos',
 'Miguel Ángel de Quevedo',
 'Copilco',
 'Universidad'])
nx.draw(ruta,with_labels= True)
```
